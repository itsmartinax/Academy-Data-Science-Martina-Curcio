# Esercizio Pandas

Create un DataFrame sulla base di un elenco di almeno 10 clienti forniti dall'utente tramite input e salvatelo in formato csv.
Per ogni cliente verranno forniti:

Numero Cliente, Nome, Cognome, Data di Nascita, Regione di Residenza.

```python
clienti={'Nome':[],'Cognome':[],'Data di Nascita':[],'Regione di Residenza':[]}
```

```python
for i in range(1):
    nome=input("Inserisci nome: ")
    clienti['Nome'].append(nome)
    cognome=input("Inserisci cognome: ")
    clienti['Cognome'].append(cognome)
    data_nascita=int(input("Inserisci anno di nascita: "))
    clienti['Data di Nascita'].append(data_nascita)
    reg_residenza=input("Inserisci regione di residenza: ")
    clienti['Regione di Residenza'].append(reg_residenza)
```

```python
import pandas as pd
```

```python
df_clienti=pd.DataFrame(clienti)
```

```python
df_clienti.to_csv('clienti.csv', index=False)
```

Partendo dal dataframe creato nell'esercizio precedente, aggiungete 3 colonne:

* La prima contiene una X in corrispondenza dei clienti minorenni;

* La seconda contiene una X in corrispondenza dei clienti con età maggiore di 17 anni;

* La terza contiene una X in corrispondenza dei clienti con età maggiore di 20 anni.

```python
import datetime as dt
```

```python
today = dt.datetime.now()
età=[]
for i in df_clienti["Data di Nascita"]:
    i=str(i)
    lista_età=today.year-int(i[-4:])
    età.append(lista_età)
```

```python
df_clienti['Età']=età
```

```python
df_clienti
```

```python
df_clienti['Minori di 18 anni'] = df_clienti['Età'].apply(lambda x : "X" if x <18 else None)
```

```python
df_clienti['Maggiori di 17 anni'] = df_clienti['Età'].apply(lambda x : "X" if x >17 else None)
```

```python
df_clienti['Maggiori di 20 anni'] = df_clienti['Età'].apply(lambda x : "X" if x > 20 else None)
```
