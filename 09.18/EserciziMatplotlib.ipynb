# **Esercizi Matplotlib**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## **Esercizio 1**

Hai a disposizione un dataset, che devi autogenerare, contenuto in un DataFrame pandas con una singola colonna temperature che rappresenta la temperatura giornaliera in una citt√† per un mese.


Scrivi un programma Python che calcoli e stampi le seguenti statistiche:
- La temperatura massima
- La temperatura minima
- La temperatura media
- La mediana delle temperature

```python
temperatura = np.random.randint(10, 31, size=31)
```

```python
dataset = {'Giorni':np.arange(1,32),'Temperatura':temperatura}
df=pd.DataFrame(dataset)
```

```python
df.head()
```

```python
temperatura_massima = np.max(df['Temperatura'])
temperatura_minima = np.min(df['Temperatura'])
temperatura_media = np.mean(df['Temperatura'])
temperatura_mediana = np.median(df['Temperatura'])
```

```python
print(f"Temperatura massima: {temperatura_massima}")
print(f"Temperatura minima: {temperatura_minima}")
print(f"Temperatura media: {temperatura_media}")
print(f"Mediana delle temperature: {temperatura_mediana}")
```

```python
plt.figure()
plt.plot(df['Giorni'],df['Temperatura'],color='red')
plt.title('Andamento della temperatura in un mese')
plt.xlabel('Giorni')
plt.ylabel('Temperatura')
plt.show()
```

```python
plt.figure()
plt.bar(df['Giorni'],df['Temperatura'], color='green')
plt.title('Andamento della temperatura in un mese')
plt.xlabel('Giorni')
plt.ylabel('Temperatura')
plt.show()
```

```python
plt.figure()
plt.scatter(df['Giorni'],df['Temperatura'], color='orange')
plt.title('Andamento della temperatura in un mese')
plt.xlabel('Giorni')
plt.ylabel('Temperatura')
plt.show()
```
